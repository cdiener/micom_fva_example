# FVA from a reference solution

Here we use a different strategy where we use the previous uptake fluxes from a MICOM result and run an FVA on the exchanges. 

This will impose an upper bound equal to the import from the reference solution, but can tell you if there is flexibility in this uptake.

```python
from micom.workflows import load_results

res = load_results("results.zip")
```

We will now use the results a basis for running FVA with the fixed uptakes from the solution.

```python
from fva import eFVA

results = eFVA(
    results=res,
    model_db="agora103_refseq216_genus_1.qza",
    taxa=["Anaerostipes"],
    fva_threshold=0.99,
    threads=4
)
```
