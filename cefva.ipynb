## FVA on MICOM models

This is quick example code on how to perform FVA for the exchange reactions for a taxon of interest in a set of existing models. 

We start by reading in some models.

Now we also wrap it into a workflow.

```python
from micom.qiime_formats import load_qiime_medium
import pandas as pd

medium = load_qiime_medium("western_diet_gut_agora.qza")
manifest = pd.read_csv("models/manifest.csv")
manifest
```

Now we run an FVA for a set of exchange reactions and taxa. Here we specify the exchnages based on their metabolite IDs. For AGORA those are usually postfixed with `[e]`.

```python
from fva import ceFVA

results = ceFVA(
    manifest=manifest,
    model_folder="models",
    medium=medium,
    metabolites=["succ[e]", "but[e]", "trp_L[e]", "glc_D[e]"],
    taxa=["Anaerostipes"],
    tradeoff=0.5,
    fva_threshold=0.95,
    strategy="minimal imports",
    threads=8
)
```
